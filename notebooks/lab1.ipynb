```python
import numpy as np

from src.visualization import plot_result
from src.algorithms import RandomTSP, NearestNeighbourTSP, GreedyCycleTSP
```

# Instance A

```python
nodes_path_A = '../data/TSPA.csv'
nodes_A = np.genfromtxt(nodes_path_A, dtype=int, delimiter=';')
```

```python
randomTSP_A = RandomTSP(nodes_path_A)
min_random_A, max_random_A, avg_random_A, edges_random_A, start_random_A = randomTSP_A.run_experiment()
```

```python
nnTSP_A = NearestNeighbourTSP(nodes_path_A)
min_nn_A, max_nn_A, avg_nn_A, edges_nn_A, start_nn_A = nnTSP_A.run_experiment()
```

```python
gcTSP_A = GreedyCycleTSP(nodes_path_A)
min_gc_A, max_gc_A, avg_gc_A, edges_gc_A, start_gc_A = gcTSP_A.run_experiment()
```

```python
def pretty_print(algorithm, minv, maxv, avgv):
    print(f'Algorithm: {algorithm}:')
    print(f'    minimal objective: {minv}')
    print(f'    mean objective: {avgv}')
    print(f'    maximal objective: {maxv}')
    print('========================================================')

pretty_print('Random', min_random_A, max_random_A, avg_random_A)
pretty_print('Nearest Neighbour', min_nn_A, max_nn_A, avg_nn_A)
pretty_print('Greedy Cycle', min_gc_A, max_gc_A, avg_gc_A)
```

```python
plot_result(nodes_A, edges_random_A, min_random_A)
```

```python
plot_result(nodes_A, edges_nn_A, min_nn_A)
```

```python
plot_result(nodes_A, edges_gc_A, min_gc_A)
```
