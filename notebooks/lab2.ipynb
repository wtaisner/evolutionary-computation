```python
import numpy as np

from src.visualization import plot_result
from src.algorithms import GreedyCycleRegretTSP
```

```python
def pretty_print(algorithm, minv, maxv, avgv):
    print(f'Algorithm: {algorithm}:')
    print(f'    minimal objective: {minv}')
    print(f'    mean objective: {avgv}')
    print(f'    maximal objective: {maxv}')
    print('========================================================')
```

# Instance A

```python
nodes_path_A = '../data/TSPA.csv'
nodes_A = np.genfromtxt(nodes_path_A, dtype=int, delimiter=';')
```

```python
gcrTSP_A = GreedyCycleRegretTSP(nodes_path_A, k=2, weight=1)
min_gcr_A, max_gcr_A, avg_gcr_A, edges_gcr_A, start_gcr_A = gcrTSP_A.run_experiment()
```

```python
gcwTSP_A = GreedyCycleRegretTSP(nodes_path_A, k=2, weight=0.5)
min_gcw_A, max_gcw_A, avg_gcw_A, edges_gcw_A, start_gcw_A = gcwTSP_A.run_experiment()
```

```python
pretty_print('Greedy Cycle', min_gcr_A, max_gcr_A, avg_gcr_A)
pretty_print('Greedy Cycle', min_gcw_A, max_gcw_A, avg_gcw_A)
```

```python
plot_result(nodes_A, edges_gcr_A, min_gcr_A)
```

```python
plot_result(nodes_A, edges_gcw_A, min_gcw_A)
```
